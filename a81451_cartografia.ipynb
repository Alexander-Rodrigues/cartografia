# Formatação dos dados

Utilizou-se um script de python para extrair os últimos n dias, no nosso caso 200, de uma base de dados de jogos. Isto traduz para meados de outubro até finais de abril. Os jogos escolhidos foram os mais populares na plataforma Steam no dia 29 de abril.
Depois do script corrido usou-se o excell para compactar os dias em semanas.
Guardou-se valores de jogadores e viewers.

```python
import pandas as pd

path = r''

df = {'csgo': pd.read_csv(path + 'csgo(X).csv'),
     'dota2': pd.read_csv(path + 'dota2(X).csv'),
     'gta5': pd.read_csv(path + 'gta5(X).csv'),
     'fm20': pd.read_csv(path + 'fm20(X).csv'),
     'tf2': pd.read_csv(path + 'tf2(X).csv')}

df['csgo'].head()
```

Os jogos usados apresentados são:

- csgo - [Counter-Strike: Global Offensive](https://store.steampowered.com/app/730/CounterStrike_Global_Offensive/)
- dota2 - [Dota 2](https://store.steampowered.com/app/570/Dota_2/)
- gta5 - [Grand Theft Auto V](https://store.steampowered.com/app/271590/Grand_Theft_Auto_V/)
- fm20 - [Football Manager 2020](https://store.steampowered.com/app/1100600/Football_Manager_2020/)
- tf2 - [Team Fortress 2](https://store.steampowered.com/app/440/Team_Fortress_2/)

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [15, 5]

x = df['csgo'].date
for key,values in df.items():
    plt.plot(x,values.players, label = key)
    
plt.xticks(rotation=45, ha="right")
plt.legend()
plt.show()
```

Este gráfico mostra que para algun jogos, especialmente com uma base de jogadores mais elevada, houve um aumento de jogadores, desde que as quarentenas começaram.
As datas estão em YY-MM-DD com uma semana entre elas, pois durante uma semana existe sempre um pico no fim de semana, logo fez-se a média de cada semana.

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [15, 5]

x = df['csgo'].date
for key,values in df.items():
    plt.plot(x,values.viewers, label = key)
    
plt.xticks(rotation=45, ha="right")
plt.legend()
plt.show()
```

A célula acima é o mesmo gráfico mas com viewers em vez de jogadores, mesmo com a média semanal estes valores são bastante voláteis com um ligeiro aumento.
